```python
## 一起分析
from pprint import pprint

nss = {}

for tar in ['edu', 'gov', 'other']:
    with open("/root/Workspace/DWTF/domain_relation/data/%s.domain_ns_info.txt" % tar,'r') as inf:
        for line in inf.readlines():
                line = line.strip()
                try:
                    domain, ns = line.split("\t")
                except:
                    print(line)
                    continue
                ns = ns.rstrip(".")
                # if this is some true ns and not dead end, add to dict.
                if ns != "~NO~NS~":
                    ds = ns.split('.')
                    try:
                        pns = ds[-2]+'.'+ds[-1]
                        if pns not in nss:
                            nss[pns] = 0
                        nss[pns] += 1
                    except Exception as e:
                        print(e, ns)
pprint(sorted(nss.items(), key=lambda x: x[1], reverse=True))
```
