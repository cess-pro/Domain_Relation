```python
## 一起分析
from pprint import pprint

nss = {}

for tar in ['edu', 'gov', 'other']:
    with open("/root/Workspace/DWTF/domain_relation/data/%s.domain_ns_info.txt" % tar,'r') as inf:
        for line in inf.readlines():
                line = line.strip()
                try:
                    domain, ns = line.split("\t")
                except:
                    print(line)
                    continue
                ns = ns.rstrip(".")
                # if this is some true ns and not dead end, add to dict.
                if ns != "~NO~NS~":
                    ds = ns.split('.')
                    try:
                        pns = ds[-2]+'.'+ds[-1]
                        if pns not in nss:
                            nss[pns] = 0
                        nss[pns] += 1
                    except Exception as e:
                        print(e, ns)
pprint(sorted(nss.items(), key=lambda x: x[1], reverse=True))
```

```python
## 分析重点的AS域

import json

with open('../data/domains_a_as_register.json', 'r') as f:
    data = json.load(f)

as_data = {}

for i in data:
    if len(data[i]['AS']) > 0:
        for asn in data[i]['AS']:
            if asn not in as_data:
                as_data[asn] = 0
            as_data[asn] += 1

for i in sorted(as_data.items(), key=lambda kv: kv[1], reverse=True):
    print(i)
```
